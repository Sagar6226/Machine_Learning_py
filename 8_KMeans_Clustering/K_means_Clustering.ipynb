```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/admin/CodeBasics Machine Learning/Machine_Learning_py/8_KMeans_Clustering/income.csv')
df.head()
```

```python
#Lets plot the Age & income over a graph to see the points
plt.scatter(df['Age'],df['Income($)'])
```

```python
#specify the number of clusters you want
km = KMeans(n_clusters=3)
km
```

```python
#Lets precdict through the model
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
#Now lets plot this array in table so that we can visualize it
df['cluster'] = y_predicted
df.head()
```

```python
#Now lets plot this array in graph so that we can visualize it
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color ='green')
plt.scatter(df2.Age,df2['Income($)'],color ='red')
plt.scatter(df3.Age,df3['Income($)'],color ='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
#red cluster is good but blue and green does not looks good this is because feature are not scaled properly we will use minmaxscaler

scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df[['Income($)']] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df[['Age']] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df.drop(['cluster'],axis='columns')
```

```python
df['cluster'] = y_predicted
df
```

```python
#Now lets plot this array in graph so that we can visualize it
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color ='green')
plt.scatter(df2.Age,df2['Income($)'],color ='red')
plt.scatter(df3.Age,df3['Income($)'],color ='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='yellow')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
#Now you can see the clusters are nicely formed
#centroids
km.cluster_centers_
```
